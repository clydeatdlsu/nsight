```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
```

```python
im_rgb = cv2.imread('night_002.png')
plt.imshow(cv2.cvtColor(im_rgb, cv2.COLOR_BGR2RGB))
```

```python
im_lab = cv2.cvtColor(im_rgb, cv2.COLOR_BGR2LAB)
#plt.imshow(cv2.cvtColor(im_lab, cv2.COLOR_BGR2LAB))
plt.imshow(im_lab)
```

```python
im_mp = mpimg.imread('night_002.png')
plt.imshow(im_mp)
```

```python
im_lab = cv2.cvtColor(im_mp, cv2.COLOR_RGB2LAB)
plt.imshow(im_lab)
```

```python
im_ycc = cv2.cvtColor(im_mp, cv2.COLOR_RGB2YCrCb)
plt.imshow(im_ycc)
```

```python
im_hsv = cv2.cvtColor(im_mp, cv2.COLOR_RGB2HSV)
plt.imshow(im_hsv)
```
